## Standard Deviation

```python
num_std = 1
```

```python
plus_1pct_std = (ds_dif['tas'].sel(experiment = '1pct').weighted(utils.onepct_model_weights).mean(dim = 'model') + 
                    num_std*ds_dif['tas'].sel(experiment = '1pct').weighted(utils.onepct_model_weights).std(dim = 'model'))
minus_1pct_std = (ds_dif['tas'].sel(experiment = '1pct').weighted(utils.onepct_model_weights).mean(dim = 'model') - 
                    num_std*ds_dif['tas'].sel(experiment = '1pct').weighted(utils.onepct_model_weights).std(dim = 'model'))
ds_1pct_compare = conv_ds.sel(experiment = '1pct').weighted(utils.onepct_model_weights).mean(dim = 'model').mean(dim = 'pulse_type')
```

```python
plus_1000gtc_std = (ds_dif['tas'].sel(experiment = '1000gtc').weighted(utils.onepct_1000gtc_weights).mean(dim = 'model') + 
                    num_std*ds_dif['tas'].sel(experiment = '1000gtc').weighted(utils.onepct_1000gtc_weights).std(dim = 'model'))
minus_1000gtc_std = (ds_dif['tas'].sel(experiment = '1000gtc').weighted(utils.onepct_1000gtc_weights).mean(dim = 'model') - 
                    num_std*ds_dif['tas'].sel(experiment = '1000gtc').weighted(utils.onepct_1000gtc_weights).std(dim = 'model'))
ds_1000gtc_compare = conv_ds.sel(experiment = '1000gtc').weighted(utils.onepct_1000gtc_weights).mean(dim = 'model').mean(dim = 'pulse_type')
```

```python
plus_std_hist = ds_hist_dif['tas'].mean(dim = 'model') + num_std*ds_hist_dif['tas'].std(dim = 'model')
minus_std_hist = ds_hist_dif['tas'].mean(dim = 'model') - num_std*ds_hist_dif['tas'].std(dim = 'model')
ds_compare_hist = conv_hist_mod_mean_ds.mean(dim = 'model').mean(dim = 'pulse_type')
```

```python
plus_std_4x = ds_4x_dif['tas'].weighted(utils.fourx_model_weights).mean(dim = 'model') + num_std*ds_4x_dif['tas'].weighted(utils.fourx_model_weights).std(dim = 'model')
minus_std_4x = ds_4x_dif['tas'].weighted(utils.fourx_model_weights).mean(dim = 'model') - num_std*ds_4x_dif['tas'].weighted(utils.fourx_model_weights).std(dim = 'model')
ds_compare_4x = conv_4x_ds.weighted(utils.fourx_model_weights).mean(dim = 'model').mean(dim = 'pulse_type')
```
