## Data

Data for this is from https://gmd.copernicus.org/articles/11/1133/2018/ CDRMIP data, where pi-CO2pulse is the 100GtC pulse and piControl is the control

```python
G_ds = xr.open_dataset('Outputs/G_internal_var_ds.nc4')['__xarray_dataarray_variable__']
G_cdr_ds = xr.open_dataset('Outputs/G_cdr_internal_var_ds.nc4')['__xarray_dataarray_variable__']
```

```python
G_ds = xr.concat([G_ds, -G_cdr_ds], pd.Index(['pulse','cdr'], name = 'pulse_type'))
G_ds.name = 'G[tas]'
```

```python
A = utils.A
```

```python
mean_G = G_ds.weighted(A).mean(dim = ['lat','lon']).mean(dim = 'pulse_type')
```

```python
model_color = utils.model_color
```

```python
internal_var = mean_G.var(dim = ['pulse_year']).mean(dim = 'model')
```

```python
model_var = mean_G.var(dim = ['model']).mean(dim = 'pulse_year')
```

```python
total_var = internal_var + model_var
```

```python
fig, ax = plt.subplots(figsize = [9,6])
plt.stackplot(model_var['year'], 
              model_var/total_var,
              internal_var/total_var,
              labels = ['Model Variance', 'Internal Variance'],
              colors = ['cadetblue','sandybrown'],
             alpha = 0.8);
plt.legend(loc = 'lower right',fontsize = 16)
plt.xlim(0,90)
plt.ylim(0,1)
plt.ylabel('Percent of total Variance', fontsize = 16)
plt.xlabel('Year of Green\'s Function', fontsize = 16)
ax.tick_params(axis='y', labelsize=16)
ax.tick_params(axis='x', labelsize=16)
plt.savefig('figures/paper/supp_variance_model_intern_GF.png', bbox_inches = 'tight', dpi = 250)
```

```python
internal_var_map = G_ds.mean(dim = 'pulse_type').var(dim = ['pulse_year']).mean(dim = 'model')
```

```python
model_var_map = G_ds.mean(dim = 'pulse_type').var(dim = ['model']).mean(dim = 'pulse_year')
```

```python
total_var_map = internal_var_map + model_var_map
```

```python
cmap = 'Reds'
levels = np.arange(0,1,0.1)
vmin = -2
vmax = 2
yr_dif = 5
fig, ((ax1, ax2),(ax3, ax4))= plt.subplots(2, 2, figsize = [15,8], subplot_kw = {'projection':ccrs.PlateCarree()}, constrained_layout = True)
plt.subplots_adjust(hspace=0.3)

########## Model Variance ##################
##contours of convolution-1pct difference
for t, ax in zip([20,85],[ax1,ax2]):
    if ax == ax1:
        add_cbar = False
    elif ax == ax2:
        add_cbar = True
    axp1 = (model_var_map/total_var_map).sel(year = slice(t-yr_dif, t+yr_dif)).mean(dim = 'year').plot(ax = ax, 
                                                                                                                  cmap = cmap, levels = levels, extend = 'both', 
                                                                                                                  add_colorbar = False,     
                                                                                                                  transform = ccrs.PlateCarree())
    ax.coastlines()
    ax.set_title(f'{t} ($\pm 5$) years', fontsize = 14)

#     if t == 20:
#         cbaxes = fig.add_axes([.49,.1,.015,.8])
#         cb = plt.colorbar(axp1, cax = cbaxes)
#         cb.set_label('$\Delta \overline{T}$ ($\degree$C)')


ax2.set_title('85 ($\pm 5$) years', fontsize = 14)


############ Internal Variance ##############
for t, ax in zip([20,85],[ax3,ax4]):
    if ax == ax3:
        add_cbar = False
    elif ax == ax4:
        add_cbar = True
    axp = (internal_var_map/total_var_map).sel(year = slice(t-yr_dif, t+yr_dif)).mean(dim = 'year').plot(ax = ax, 
                                                                                                                  cmap = cmap, levels = levels, extend = 'both', 
                                                                                                                  add_colorbar = False,     
                                                                                                                  transform = ccrs.PlateCarree())

    ax.coastlines()
    

cbaxes = fig.add_axes([0.92,.1,.015,.8])
cb = plt.colorbar(axp, cax = cbaxes)
cb.set_label('% of total variance')

ax1.text(-0.05,.4,'Model', rotation = 'vertical', fontsize = 14, transform=ax1.transAxes)
ax3.text(-0.05,.25,'Internal Variability', rotation = 'vertical', fontsize = 14, transform=ax3.transAxes)
ax3.set_title(None)
ax4.set_title(None)

#subplot labels, a, b, c, d
alpha_labels = iter(alphabet)
for ax in [ax1, ax2, ax3, ax4]:
    label = next(alpha_labels)
    ax.text(x = -0.07, y =1.1, s = label, transform=ax.transAxes, fontweight="bold", fontsize = 14)

plt.savefig(f'figures/paper/supp_map_variance.png', bbox_inches = 'tight', dpi = 300)
```

```python
mean_intramod = G_ds.weighted(A).mean(dim = ['lat','lon']).sel(pulse_year = 0).mean(dim = 'pulse_type').mean(dim = 'model').load()
std_intramod = G_ds.weighted(A).mean(dim = ['lat','lon']).sel(pulse_year = 0).mean(dim = 'pulse_type').std(dim = 'model').load()
```

```python
mean_intern = {}
for m in  G_ds.model.values:
    mean_intern[m] = G_ds.mean(dim = 'pulse_type').weighted(A).mean(dim = ['lat','lon']).sel(model = m).mean(dim = 'pulse_year').load()
std_intern = {}
for m in  G_ds.model.values:
    std_intern[m] = G_ds.mean(dim = 'pulse_type').weighted(A).mean(dim = ['lat','lon']).sel(model = m).std(dim = 'pulse_year').load()
    
```

```python
mean_intramod_type = G_ds.weighted(A).mean(dim = ['lat','lon']).sel(pulse_year = 0).mean(dim = 'model').load()
std_intramod_type = G_ds.weighted(A).mean(dim = ['lat','lon']).sel(pulse_year = 0).std(dim = 'model').load()
```

```python
mean_intern_type = {}
for m in  G_ds.model.values:
    mean_intern_type[m] = G_ds.weighted(A).mean(dim = ['lat','lon']).sel(model = m).mean(dim = 'pulse_year').load()
std_intern_type = {}
for m in  G_ds.model.values:
    std_intern_type[m] = G_ds.weighted(A).mean(dim = ['lat','lon']).sel(model = m).std(dim = 'pulse_year').load()
    
```
