```python
G_ds = xr.open_dataset('Outputs/G_ds.nc4')['__xarray_dataarray_variable__']

G_CDR_ds = xr.open_dataset('Outputs/G_cdr_ds.nc4')['__xarray_dataarray_variable__']

#4th order polyfit
Gpoly = G_ds.polyfit('year', 4)
G_ds= xr.polyval(G_ds.year, Gpoly)['polyfit_coefficients']

Gpoly_cdr = G_CDR_ds.polyfit('year', 4)
G_CDR_ds= xr.polyval(G_CDR_ds.year, Gpoly_cdr)['polyfit_coefficients']
```

## Pattern of GHG only vs. CO2

```python
# additional potential useful resource for emissions/concentration data: https://esgf-node.llnl.gov/search/input4mips/
```

```python
## import the emissions of non-co2 GHG's and turn into a dataset
## data is from: https://tntcat.iiasa.ac.at/SspDb/dsd?Action=htmlpage&page=welcome

ghg_list = ['C2F6', 'CH4', 'SF6', 'N2O', 'HFC', 'CF4']
ds = {}
for ghg in ghg_list:
    df = pd.read_excel(f'cmip6_data/ssp245_emissions/{ghg}_total.xlsx')
    ds[ghg] = df.iloc[0, 5:-1].to_xarray().rename({'index':'s'})
    ds[ghg].attrs = {'units':df['Unit'][0]}
    
for ghg in ['CH4', 'HFC']: #some are in Mt, switch to gt
    ds[ghg] *= 100
    
ds_emis = xr.concat([ds[m] for m in ds.keys()], pd.Index([m for m in ds.keys()], name='ghg'))
#ds_emis['s'] = (ds_emis['s'] - ds_emis['s'][0]).astype('int')  # put into timeline that matches our SSP (year 0 = 2015)
ds_emis.attrs = {'units': 'kt/yr'}
ds_emis = ds_emis.astype('float64')
ds_emis['s'] = ds_emis['s'].astype('int')

#interpolate for yearly emissions
ds_emis = ds_emis.interp(s = np.arange(2015,2101))
```

```python
#estimate historical emissions as linear growth from zero in 1850
ds_hist_emis = xr.DataArray(
                        data = np.ones([len(ds_emis['ghg']), len(np.arange(1850,2015))]),
                        dims = ['ghg', 's'],
                        coords = dict(ghg = (['ghg'], ds_emis['ghg'].values),
                                     s = (['s'], np.arange(1850,2015)))
                            )

for ghg in ghg_list:
    ds_hist_emis.loc[dict(ghg = ghg)] = np.linspace(0,ds_emis.sel(s = 2015, ghg = ghg).values, (2015-1850))
```

```python
ds_emis = xr.concat([ds_hist_emis, ds_emis], dim = 's') #merge the historical and future emissions
ds_emis['s'] = ds_emis['s'] - 2015 #set the time so that 2015 is year 0 like in our temperature profiles
```
